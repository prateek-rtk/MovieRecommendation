```python
import numpy as np
import pandas as pd
import ast
```

```python
m0vies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
m0vies.head(5)
```

```python
m0vies.tail(5)
```

```python
credits.head(5)
```

```python
credits.tail(5)
```

```python
m0vies=m0vies.merge(credits, on='title')
```

```python
m0vies.head(1)
```

```python
#extracting feature columns- genres, id, keywords, title, overview, cast, crew
m0vies=m0vies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
m0vies.head()
```

```python
m0vies.isnull().sum()
```

```python
m0vies.dropna(inplace=True)
m0vies.isnull().sum()
```

```python
m0vies.duplicated().sum()
```

```python
m0vies.iloc[0].genres
```

```python
#change string format to list format
def change(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
m0vies['genres'] = m0vies['genres'].apply(change)
```

```python
m0vies.head(5)
```

```python
m0vies['keywords']=m0vies['keywords'].apply(change)
```
